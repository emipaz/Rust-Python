# Iterar Vectores en Rust con `for`

En este video se explica cómo **iterar un vector en Rust usando un bucle `for`**, y se profundiza en las diferencias fundamentales respecto a otros lenguajes debido al sistema de **ownership (propiedad)** y **borrowing (préstamo)**.

## 1️⃣ Iteración básica de un vector

Dado un vector:

````rust
let numeros = vec![1, 2, 3];

for n in numeros {
    println!("{}", n);
}
````

⚠️ Problema: Transferencia de propiedad (Move)

Si después del for intentamos hacer:
```rust
println!("{:?}", numeros);
```
Obtendremos un error:


````rust
let numeros = vec![1, 2, 3];

for n in numeros {
    println!("{}", n);
}

println!("{:?}", numeros);
````

> borrow of moved value

### ¿Por qué ocurre esto?

En Rust, la variable numeros es la propietaria (owner) del vector.

Al usar:

```rust
for n in numeros
```

Se crea internamente un iterador.

La propiedad del vector se transfiere (move) al iterador.

numeros deja de ser válido.

Ya no podemos acceder a él después del bucle.

- En otros lenguajes (Java, C#, Python):
  - No existe el concepto de ownership.
  - Se trabaja con referencias.
  - No se transfiere la propiedad.
  - El objeto sigue siendo accesible tras el bucle.

- Rust es más estricto porque:
  - No tiene garbage collector.
  - Gestiona memoria mediante ownership.


### 2️⃣ Solución: Usar referencias (Borrowing)

Para evitar la transferencia de propiedad, usamos una referencia:

````rust

let numeros = vec![1, 2, 3];

for n in &numeros {
    println!("{}", n);
}

println!("{:?}", numeros);
````

> Ahora:

- Se realiza un préstamo (borrow).
- numeros sigue siendo el propietario.

Podemos usar el vector después del bucle sin errores.


### 3️⃣ Modificar los valores del vector

Si queremos modificar los elementos mientras iteramos, debemos:

- Paso 1: 
  - Hacer el vector mutable
    ```rust
    let mut numeros = vec![1, 2, 3];
    ```
- Paso 2: Usar referencia mutable en el for
    
    ```rust
    for n in &mut numeros {
    *n += 1;
    }   
    ```

- ¿Por qué usar *n?
  - n es una referencia mutable.
- No podemos hacer:
  ```rust
  n += 1; // ❌ Error
  ```

Porque es una referencia.

Debemos desreferenciar con *:
```rust
*n += 1; // ✅ Correcto
```

El * es el operador de desreferenciación.


````rust
let mut numeros = vec![1, 2, 3];

for n in &mut numeros {
    
    *n *= 2;
}

println!("{:?}", numeros);
````

## 📌 Resumen Final

- Iteración normal (for n in numeros)
  - Transfiere la propiedad (move).
  - La variable original queda invalidada.

- Iteración con referencia (for n in &numeros)
  - Se usa borrowing.
  - No se transfiere la propiedad.
  - Se puede usar el vector después.

- Modificación de elementos
  - El vector debe ser mut.
  - Se usa &mut en el for.
  - Se desreferencia con * para modificar el valor.

Aunque el sistema de ownership puede parecer complejo al principio, es fundamental para entender cómo Rust gestiona la memoria de forma segura y eficiente sin necesidad de garbage collector.
